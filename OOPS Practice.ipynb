#### User login Authentication

```python
class Login: # Parent Class
    def __init__(self):  # Constructor/intializer
        self.name = None # Attributes/properties
        self.password = None # Attributes/properties

    def user_auth(self): # Method
        if self.name == "admin" and self.password == "admin":
            return f"User {self.name} has successfully logged in."
        else:
            return "Authentication failed."
        
# Create an instance of the `Login` class.
l1 = Login()

# Prompt the user to enter their username and password.
l1.name = input("User Name: ")
l1.password = input("Password: ")

# Call the `user_auth` method to perform user authentication and store the result.
result = l1.user_auth()

# Display the authentication result.
print(result)
    
    
```

#### Customer Account Creation

```python
class Accounts(): # Parent Class
    def __init__(self): # Constructor/intializer
        self.account_name = None  # Attributes/properties
        self.account_site = None 
        self.account_address = None
        self.account_address_line_2 = None
        self.city = None
        self.zip_code = None
        self.country = None
        self.main_phone_number = None
        self.main_fax_number = None
        self.url = None  # Attributes/properties
    
    def display(self): # Method
        return f"""
        Accounts Name: {self.account_name}
        Account Site Name: {self.account_site}
        Account Address: {self.account_address}
        Account Address Line 2: {self.account_address_line_2}
        City: {self.city}
        Zip Code: {self.zip_code}
        Country: {self.country}
        Main Phone Number: {self.main_phone_number}
        Fax Number: {self.main_fax_number}
        URL: {self.url}
        """
# Create an instance of the `Accounts` class.
A1 = Accounts()

# Prompt the user to input account information.
A1.account_name = input("Account Name : ")
A1.account_site = input("Account Site Name : ")
A1.account_address = input("Account Address : ")
A1.account_address_line_2 = input("Account Address Line 2 : ")
A1.city = input("City : ")
A1.zip_code = input("Zip Code : ")
A1.country = input("Country : ")
A1.main_phone_number = input("Main Phone Number : ")
A1.main_fax_number = input("Fax Number : ")
A1.url = input("URL : ")

# Call the `display` method to format and display the entered account information.
print (A1.display())
```

#### Customer Asset Creation

```python
class assets(Accounts): # Child Class
    def __init__(self): # Constructor/intializer
        super().__init__() # Call the parent class constructor
        self.assets_number = None # Attributes/properties
        self.serial_number = None
        self.asset_category = None
        self.product = None
        self.status = None
        self.service_status = None
        self.operating_status = None
        self.warranty_start_date = None
        self.warrnat_end_date = None
        self.agreement = None # Attributes/properties
    
    def display_assets(self): # Method
        return f"""
        Account : {self.account_name}
        Asset #: {self.assets_number}
        Serial #: {self.serial_number}
        Asset Category: {self.asset_category}
        Product: {self.product}
        Status: {self.status}
        Service Status: {self.service_status}
        Operating Status: {self.operating_status}
        Account: {self.account_name}
        Warranty-Start Date: {self.warranty_start_date}
        Warranty-End Date: {self.warrnat_end_date}
        Agreement: {self.agreement}

        """

AA1=assets()
AA1.account_name = A1.account_name
AA1.assets_number = input("Asset #:")
AA1.serial_number = input("Serial #:")
AA1.asset_category = input("Asset Category:")
AA1.product = input("Product:")
AA1.status = input("Status:")
AA1.service_status = input("Service Status:")
AA1.operating_status = input("Operating Status:")
AA1.warranty_start_date = input("Warranty-Start Date:")
AA1.warrnat_end_date = input("Warranty-End Date:")
AA1.agreement = input("Agreement:")

print (AA1.display_assets())
```

#### Customer Product Creation

```python
class product(Accounts): # Child Class
    def __init__(self): # Constructor/intializer
        self.product_name = None # Attributes/properties
        self.part_number = None
        self.description = None
        self.status = None # Attributes/properties
        
    def display_product(self): # Method
        return f"""
        Product: {self.product_name}
        Part #: {self.part_number}
        Description: {self.description}
        Status: {self.status}
        
        """
P1 = product()
P1.product_name = input("Product")
P1.part_number = input("Part #")
P1.description = input("Description:")
P1.status = input("Status:")

print (P1.display_product())
```

#### Customer Contacts Creation

```python
class contacts(Accounts): # Child Class
    def __init__(self): # Constructor/intializer
        super().__init__() # Call the parent class constructor
        self.fname = None # Attributes/properties
        self.Lname = None
        self.emailaddress = None
        self.job_title = None
        #self.account = None
        self.work_phone_number = None
        self.mobile_phone_number = None
        self.account_address = None # Attributes/properties
    
    def Display_contact(self): # Method
        return f"""
        First Name: {self.fname}
        Last Name: {self.Lname}
        Email: {self.emailaddress}
        Job Title: {self.job_title}
        Account Name: {self.account_name}
        Work Phone #: {self.work_phone_number}
        Mobile Phone #: {self.mobile_phone_number}
        Account Address: {self.account_address}

        """
C1 = contacts()
C1.account_name = A1.account_name
C1.fname = input("First Name")
C1.Lname = input("Last Name:")
C1.emailaddress = input("Email:")
C1.job_title = input("Job Title:")
#C1.account = input("Account Name:")
C1.work_phone_number = input("Work Phone #:")
C1.mobile_phone_number = input("Mobile Phone #:")
C1.account_address = A1.account_address

print (C1.Display_contact())
```

#### Customer Ticket Generation and Assigned to Engineer

```python
class ticketgen(Accounts): # Child Class
    def __init__(self): # Constructor/intializer
        super().__init__() # Call the parent class constructor
        self.service_request = None # Attributes/properties
        self.Summary = None
        self.description = None
        self.area_type = None
        self.sub_area_issue = None
        self.source = None
        self.service_group = None
        self.status = None
        self.substatus = None
        self.owner = None
        self.incident_report_date = None # Attributes/properties
        
    def display_ticketgen(self): # Method
        return f"""
        SR #: {self.service_request}
        Last Name : {self.Lname}
        First Name : {self.fname}
        Account Name : {self.account_name}
        Site Name : {self.account_site}
        City : {self.city}
        Product : {self.product_name}
        Asset #: {self.assets_number}
        Serial #:{self.serial_number}
        Summary: {self.Summary}
        Description:{self.description}
        Area / Type:{self.area_type}
        Subarea / Issue:{self.sub_area_issue}
        Source:{self.source}
        Service Group:{self.service_group}
        Status:{self.status}
        Substatus:{self.substatus}
        Owner:{self.owner}
        Incident Report Date:{self.incident_report_date}
        
        Ticket has been generated and assinged to Mr{self.owner}
        """
DT1 = ticketgen()
DT1.service_request = "12345678910"
DT1.Lname = C1.Lname
DT1.fname = C1.fname
DT1.account_name = A1.account_name
DT1.account_site = A1.account_site
DT1.city = A1.city
DT1.product_name = P1.product_name
DT1.assets_number = AA1.assets_number
DT1.serial_number = AA1.serial_number
DT1.Summary = input("Summary:")
DT1.description = input("Description:")
DT1.area_type = input("Area / Type:")
DT1.sub_area_issue = input("Subarea / Issue:")
DT1.source = input("Source:")
DT1.service_group = input("Service Group:")
DT1.status = input("Status:")
DT1.substatus = input("Substatus:")
DT1.owner = input("Owner:")
DT1.incident_report_date = input("Incident Report Date:")

print (DT1.display_ticketgen())
```
